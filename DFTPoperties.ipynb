```python
import numpy as np
import matplotlib.pyplot as plt
n = np.linspace(-10, 10, 500)
x = np.sinc(n*np.pi)
plt.plot(n, x)
plt.title("Sinc Function")
plt.ylabel("Sinc(n)")
plt.xlabel("n")
plt.show()
```

```python
X_k = np.fft.fft(x)

plt.subplot(2,1,1)
plt.plot(np.abs(np.fft.fftshift(X_k)))

plt.ylabel("|X(k)|")
plt.xlabel("k")
plt.subplot(2,1,2)
plt.plot(np.unwrap(np.angle(np.fft.fftshift(X_k))))
plt.ylabel("Fase(X(k))")
plt.xlabel("k")
plt.show()
```

```python
shift = 10
n = np.linspace(-10, 10, 500)

x1 = np.sinc(n*np.pi - shift)
  
plt.plot(n, x1)
plt.title("Sinc Function - Shifted")
plt.ylabel("Sinc(n-10)")
plt.xlabel("n")
plt.show()
```

```python
X1_k = np.fft.fft(x1,1024)

plt.subplot(2,1,1)
plt.plot(np.abs(np.fft.fftshift(X1_k)))
plt.subplot(2,1,2)
plt.ylabel("|X1(k)|")
plt.xlabel("k")
plt.plot(np.unwrap(np.angle(np.fft.fftshift(X1_k))))
plt.ylabel("Phase(X(k))")
plt.xlabel("k")
plt.show()
```
