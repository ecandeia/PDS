## <font color='blue'> Amostragem </font>

**Definição:** amostragem <font color='blue'> periódica </font> é o processo que produz uma representação no tempo discreto, $x(n)$, de um sinal analógico $x_c(t)$
$$
x(n) = x_c(nT), \ \ -\infty < n < \infty
$$
sendo $T$ o <font color='blue'>período de amostragem </font>, e $F_s = \frac{1}{T}$, a <font color='blue'>frequência de amostragem</font>

Matematicamente
$$
s(t) = \sum_{n=-\infty}^{ \infty}\delta(t - nT)
$$
e
$$
x_s(t) = x_c(t)s(t) = x_c(t) \sum_{n=-\infty}^{ \infty}\delta(t - nT)
$$
$$
x_s(t) = \sum_{n=-\infty}^{ \infty}x_c(nT)\delta(t - nT)
$$

Transformada de Fourier de uma sequência periódica de impulsos, $s(t)$
$$
S(j\Omega) = \frac{2\pi}{T}\sum_{k=-\infty}^{ \infty}\delta(\Omega - k\Omega_s)
$$
sendo $\Omega_s = 2\pi/T$, então como $X_s(j\Omega) = \frac{1}{2\pi}X_c(j\Omega)*S(j\Omega)$, temos
$$
X_s(j\Omega) = \frac{1}{T} \sum_{k=-\infty}^{ \infty}X_c[j(\Omega - k\Omega_s)]
$$

![title](img/fig43aOppenheim2aEdicao.png)


Para facilitar nossas análises, vamos considerar que $x_c(t)$ é um sinal limitado em banda, ou seja, $X_c(j\Omega) = 0$, para $|\Omega| > \Omega_N$.

***Teorema de Nyquist:*** Um sinal $x_c(t)$  limitado em banda, com $X_c(j\Omega) = 0$ para $|\Omega| \geq \Omega_N$, pode ser univocamente recuperado de suas amostras $x(n)$ tomadas a uma taxa de
$$
\Omega_s \geq 2\Omega_N
$$
Observação: Esse teorema também é atribuído a Shannon e Kotelnikov.


<font color='green'>Exercício:</font> Seja $x_c(t) = \cos(\Omega_0 t)$, esboce o espectro de $x_s(t)$ obtido amostrando $x_c(t)$ para $\Omega_s > 2\Omega_0$ e $\Omega_s < 2 \Omega_0$.

***Espectro Analógico x Discreto:*** Lembremos que o sinal amostrado é dado por 
$x_s(t) = \sum\limits_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)$ e aplicando Fourier
$$
X_s(j\Omega) = \int_{-\infty}^{\infty} x_s(t)e^{-j\Omega t}dt = \int_{-\infty}^{\infty} \sum_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)e^{-j\Omega t}dt
$$
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty} x_c(nT_s)\int_{-\infty}^{\infty} \delta(t - nT_s)e^{-j\Omega t}dt  = \sum_{n=-\infty}^{\infty} x_c(nT_s)e^{-j\Omega T_s}  
$$
e como 
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x(n)e^{-j\omega}
$$
comparando as duas equações temos que se fizermos $\omega = \Omega T_s$ na segunda obtemos a primeira, logo $X(e^{j\omega})|_{\omega = \Omega Ts} = X_s(j\Omega)$

Até o momento trabalhamos com o sinal $x_c(t)$ do qual obtemos 
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty}x_c(nT_s)e^{-j\Omega T_s n}
$$
entretanto, se observarmos apenas as amostras, sem considerar o período de amostragem temos
$$
x[n] = x_c(nT_s)
$$
e
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x[n]e^{-j\omega n}
$$
ou seja
$$
X_s(j\Omega) = X(e^{j\omega})|_{\omega=\Omega T_s} = X(e^{j\Omega T_s})
$$
Temos que 
$$
X(e^{j\Omega T_s}) = \frac{1}{T}\sum_{k=-\infty}^{\infty}X_c(j(\Omega - k\Omega_s))
$$
logo
$$
X(e^{j\omega}) =\frac{1}{T_s}\sum_{k=-\infty}^{\infty}X_c\left(j\left(\frac{\omega}{T_s} - \frac{2\pi k}{T_s}\right)\right)
$$


***Análise de relação entre $\Omega$ e $\omega$***

Partindo do fato que $\omega = \Omega T$, observemos que:
1. $\omega = \pi$ então $\pi = \Omega T$ e $2\pi f T = \pi$ isolando $f$ temos $ f = \frac{1}{2T} = \frac{F_s}{2}$, 

2. $\omega = 2\pi$ então $\pi = \Omega T$ que corresponde a $f = F_s$, e

3. Se $F_B$ for a maior frequência do sinal $\omega_B = 2\pi F_B T_s = \frac{2\pi F_B}{F}$ e como $F_s > 2F_B$, $\omega_B < \pi$.

A componente do sinal analógico que for mapeada em $\Omega_s - \Omega_B$ em $X_s(j\Omega)$ será mapeada em $\omega = 2\pi - \frac{2\pi F_B}{F_s}$ que terá valor $\pi < \omega < 2\pi$.

***Exemplo:*** Analisar as correspondências para um sinal $x_c(t) = \cos(\omega_B t)$.

***Reconstrução a partir das amostras***

Partindo das amostras
$$
x_s(t) = \sum_{n=-\infty}^{\infty}x(n)\delta(t - nT_s)
$$
sendo $T$ o período de amostragem, e $x(n)$ o valor da amostra em $t = nT$.

Para reconstruir o sinal analógico a partir das amostras, podemos utilizar um filtro passa-baixas ideal $H_r(j\Omega)$, e obter na sua saída
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT_s)
$$
sendo $h_r(t)$ a resposta ao impulso do filtro.

Sendo $h_r(t)$ é um filtro passa baixas ideal, o módulo de sua resposta em frequência temo o formato
![title](img/FPB.png)
cuja transformada inversa é
$$
h_r(t) = \frac{\sin(\pi t/T)}{\pi t/T}
$$
![title](img/fig48cOppenheim2aEdicao.png)


A reconstrução ser dada por
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT) = \sum_{n=-\infty}^{\infty}x[n]\frac{\sin[\pi(t -nT)/T]}{\pi(t - nT)/T},\ \ logo
$$

***Filtragem anti-aliasing***

Sinal de entrada não seja limitado em banda, ou taxa de Nyquist necessária é muito alta;

Filtrar o sinal;

Ex.: Sinal de voz, sinal da rede elétrica;

Filtro passa baixas:
$$
H(j\Omega) = \left\{\begin{array}{c c} 1, & |\Omega| < \Omega_c  \\ 0, & |\Omega| > \Omega_c \end{array}\right.
$$
