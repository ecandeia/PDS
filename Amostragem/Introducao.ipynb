## <font color='blue'> Amostragem </font>

**Definição:** amostragem <font color='blue'> periódica </font> é o processo que produz uma representação no tempo discreto, $x(n)$, de um sinal analógico $x_a(t)$
$$
x(n) = x_a(nT), \ \ -\infty < n < \infty
$$
sendo $T$ o <font color='blue'>período de amostragem </font>, e $F_s = \frac{1}{T}$, a <font color='blue'>frequência de amostragem</font>

A conversão é de um sinal no tempo contínuo para um sinal no tempo discreto. Que Conversão ideal de contínuo para discreto (C/D), que nesse momento pode ser visualizada como uma chave que fecha e abre a cada $T$ instantes de tempo, conforme representado na Figura abaixo.

![title](img/ConvContinuoDiscreto.png)

Saliente-se que a conversão realizada até esse momento é de contínuo para discreto, pois um conversor Analógico para Digital necessita de outros blocos.

Uma representação do conversor analógico para discreto pode ser feita consdeirando a multiplicação sinal analógico $x_a(t)$ por um trem de impulsos periódicos com período $T$:
![title](img/ConversaoTremimpulsoSeqDiscreta.png)


Matematicamente
$$
s(t) = \sum_{n=-\infty}^{ \infty}\delta(t - nT)
$$
e
$$
x_s(t) = x_a(t)s(t) = x_a(t) \sum_{n=-\infty}^{ \infty}\delta(t - nT)
$$
$$
x_s(t) = \sum_{n=-\infty}^{ \infty}x_a(nT)\delta(t - nT)
$$

Transformada de Fourier de uma sequência periódica de impulsos, $s(t)$
$$
S(j\Omega) = \frac{2\pi}{T}\sum_{k=-\infty}^{ \infty}\delta(\Omega - k\Omega_s)
$$
sendo $\Omega_s = 2\pi/T$, então como $X_s(j\Omega) = \frac{1}{2\pi}X_c(j\Omega)*S(j\Omega)$, temos
$$
X_s(j\Omega) = \frac{1}{T} \sum_{k=-\infty}^{ \infty}X_a[j(\Omega - k\Omega_s)]
$$

![title](img/EspectroAnalogicoeTremdeImpulsos.png)


Para facilitar nossas análises, vamos considerar que $x_c(t)$ é um sinal limitado em banda, ou seja, $X_c(j\Omega) = 0$, para $|\Omega| > \Omega_a$.

***Teorema de Nyquist:*** Um sinal $x_c(t)$  limitado em banda, com $X_a(j\Omega) = 0$ para $|\Omega| \geq \Omega_a$, pode ser univocamente recuperado de suas amostras $x(n)$ tomadas a uma taxa de
$$
\Omega_s \geq 2\Omega_a
$$
Observação: Esse teorema também é atribuído a Shannon e Kotelnikov.


<font color='green'>Exercício:</font> Seja $x_a(t) = \cos(\Omega_0 t)$, esboce o espectro de $x_s(t)$ obtido amostrando $x_a(t)$ para $\Omega_s > 2\Omega_0$ e $\Omega_s < 2 \Omega_0$.

***Espectro Analógico x Discreto:*** Lembremos que o sinal amostrado é dado por 
$x_s(t) = \sum\limits_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)$ e aplicando Fourier
$$
X_s(j\Omega) = \int_{-\infty}^{\infty} x_s(t)e^{-j\Omega t}dt = \int_{-\infty}^{\infty} \sum_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)e^{-j\Omega t}dt
$$
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty} x_c(nT_s)\int_{-\infty}^{\infty} \delta(t - nT_s)e^{-j\Omega t}dt  = \sum_{n=-\infty}^{\infty} x_c(nT_s)e^{-j\Omega T_s}  
$$
e como 
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x(n)e^{-j\omega}
$$
comparando as duas equações temos que se fizermos $\omega = \Omega T_s$ na segunda obtemos a primeira, logo $X(e^{j\omega})|_{\omega = \Omega Ts} = X_s(j\Omega)$

Até o momento trabalhamos com o sinal $x_c(t)$ do qual obtemos 
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty}x_c(nT_s)e^{-j\Omega T_s n}
$$
entretanto, se observarmos apenas as amostras, sem considerar o período de amostragem temos
$$
x[n] = x_c(nT_s)
$$
e
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x[n]e^{-j\omega n}
$$
ou seja
$$
X_s(j\Omega) = X(e^{j\omega})|_{\omega=\Omega T_s} = X(e^{j\Omega T_s})
$$
Temos que 
$$
X(e^{j\Omega T_s}) = \frac{1}{T}\sum_{k=-\infty}^{\infty}X_c(j(\Omega - k\Omega_s))
$$
logo
$$
X(e^{j\omega}) =\frac{1}{T_s}\sum_{k=-\infty}^{\infty}X_c\left(j\left(\frac{\omega}{T_s} - \frac{2\pi k}{T_s}\right)\right)
$$


***Análise de relação entre $\Omega$ e $\omega$***

Partindo do fato que $\omega = \Omega T$, observemos que:
1. $\omega = \pi$ então $\pi = \Omega T$ e $2\pi f T = \pi$ isolando $f$ temos $ f = \frac{1}{2T} = \frac{F_s}{2}$, 

2. $\omega = 2\pi$ então $\pi = \Omega T$ que corresponde a $f = F_s$, e

3. Se $F_B$ for a maior frequência do sinal $\omega_B = 2\pi F_B T_s = \frac{2\pi F_B}{F_s}$ e como $F_s > 2F_B$, $\omega_B < \pi$.

A componente do sinal analógico que for mapeada em $\Omega_s - \Omega_B$ em $X_s(j\Omega)$ será mapeada em $\omega = 2\pi - \frac{2\pi F_B}{F_s}$ que terá valor $\pi < \omega < 2\pi$.

***Exemplo:*** Analisar as correspondências para um sinal $x_c(t) = \cos(\omega_B t)$.

***Reconstrução a partir das amostras***

Partindo das amostras
$$
x_s(t) = \sum_{n=-\infty}^{\infty}x(n)\delta(t - nT_s)
$$
sendo $T_s$ o período de amostragem, e $x(n)$ o valor da amostra em $t = nT_s$.

Para reconstruir o sinal analógico a partir das amostras, podemos utilizar um filtro passa-baixas ideal $H_r(j\Omega)$, e obter na sua saída
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT_s)
$$
sendo $h_r(t)$ a resposta ao impulso do filtro passa baixas ideal, cujo módulo de resposta em frequência tem o formato
![title](img/FPB.png)
com transformada inversa
$$
h_r(t) = \frac{\sin(\pi t/T_s)}{\pi t/T_s}
$$

A reconstrução ser dada por
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT_s) = \sum_{n=-\infty}^{\infty}x[n]\frac{\sin[\pi(t -nT_s)/T_s]}{\pi(t - nT_s)/T_s},\ \ logo
$$

```python
## Demonstra a recuperação pela soma de sincs
## Contribuição de Joao Pedro Melquiades Gomes <joao.melquiades@ee.ufcg.edu.br>
## Outubro de 2022

import numpy as np
import matplotlib.pyplot as plt
# Seaborn is only to make the plot more beautiful
import seaborn as sns
sns.set_theme()

# Parameters
N = 10 # Number of samples
f1 = 3000 #Hz
f2 = 9500 #Hz
fs = 2*max(f1, f2) # Sampling frequency
n = np.arange(N)/fs #Discrete time
ta = np.linspace(0, N/fs, N*100) # Analogic time

fn = np.cos(2*np.pi*f1*n) + np.cos(2*np.pi*f2*n)
ft = np.cos(2*np.pi*f1*ta) + np.cos(2*np.pi*f2*ta)

plt.figure(figsize=(20,5))
plt.scatter(n, fn, color='red')
plt.plot(ta, ft)
plt.title('Arbitrary signal (analogic and sampled)')
plt.xlabel('Time, s')
plt.ylabel('Absolute amplitude')
plt.show()
```

```python
quit = False

recovered = 0
sinc_plots = []

skip_all = False
skip = input('Skip step-by-step?(y/n) ')

if(skip == 'y'):
    skip_all = True
for i in range(N):
        sinc = np.sinc((ta - i/fs)*fs)
        parcel = fn[i]*sinc
        recovered += parcel
        sinc_plots.append(parcel)
        
        
        if(not skip_all):
            plt.figure(figsize=(20,5))
            plt.plot(ta, recovered, linewidth=3)
            plt.scatter(n, fn, color='red')
            for snc in sinc_plots:
                plt.plot(ta, snc, '--')
            plt.show()
            input('Continue...')
```
