## <font color='blue'> Amostragem </font>

**Definição:** amostragem <font color='blue'> periódica </font> é o processo que produz uma representação no tempo discreto, $x(n)$, de um sinal analógico $x_a(t)$
$$
x(n) = x_a(nT_s), \ \ -\infty < n < \infty
$$
sendo $T_s$ o <font color='blue'>período de amostragem </font>, e $F_s = \frac{1}{T_s}$, a <font color='blue'>frequência de amostragem.</font>

A conversão é de um sinal no tempo contínuo para um sinal no tempo discreto (C/D), que nesse momento pode ser visualizada como a ação de uma chave que fecha e abre a cada $T_s$ instantes de tempo, conforme representado na Figura abaixo.

![title](img/ConvContinuoDiscreto.png)

Saliente-se que a conversão realizada até esse momento é de contínuo para discreto, pois um conversor Analógico para Digital necessita de outros blocos.

Uma representação do conversor analógico para discreto pode ser feita considerando a multiplicação do sinal analógico $x_a(t)$ por um trem de impulsos periódicos com período $T_s$.
![title](img/ConversaoTremimpulsoSeqDiscreta.png)


Matematicamente
$$
s(t) = \sum_{n=-\infty}^{ \infty}\delta(t - nT_s)
$$
e
$$
x_s(t) = x_a(t)s(t) = x_a(t) \sum_{n=-\infty}^{ \infty}\delta(t - nT_s)
$$
$$
x_s(t) = \sum_{n=-\infty}^{ \infty}x_a(nT_s)\delta(t - nT_s)
$$


```python
%pylab inline


fig,ax = subplots()

f = 1.0  # Hz, frequencia do sinal
fs = 10 # Hz, frequência de amostragem (ie. >= 2*f)
t = arange(-1,1+1/fs,1/fs) # intervalor de amostragem
x = sin(2*pi*f*t)
ax.plot(t,x,'o-')
ax.set_xlabel('tempo',fontsize=18)
ax.set_ylabel('amplitude',fontsize=18)
ax.grid()
```


Lembrando que a transformada de Fourier de uma sequência periódica de impulsos, $s(t)$
$$
S(j\Omega) = \frac{2\pi}{T_s}\sum_{k=-\infty}^{ \infty}\delta(\Omega - k\Omega_s)
$$
sendo $\Omega_s = 2\pi/T_s$, então como $X_s(j\Omega) = \frac{1}{2\pi}X_c(j\Omega)*S(j\Omega)$, temos
$$
X_s(j\Omega) = \frac{1}{T_s} \sum_{k=-\infty}^{ \infty}X_a[j(\Omega - k\Omega_s)]
$$

![title](img/EspectroAnalogicoeTremdeImpulsos.png)


Para facilitar nossas análises, vamos considerar que $x_c(t)$ é um sinal limitado em banda, ou seja, $X_c(j\Omega) = 0$, para $|\Omega| > \Omega_a$.

***Relação entre $\Omega_s$ e $\Omega_a$***

Vamos analisar o que ocorre com o espectro de fequências do sinal amostrado quando se escolhe valores para $\Omega_s$, para tanto sejam as seguintes possibilidades:

1. $\Omega_s > \Omega_a$: neste caso não há sobreposição das cópias do espectro de $X_a(j\Omega)$
![title](img/EspectrosemAliasing.png)
2. $\Omega_s < \Omega_a$: ocorre a sbreposição das cópias do espectro, fenômeno denominado de **aliasing**
![title](img/EspectrocomAliasing.png)

***Recuperação do sinal a partir das amostras***

Dado que temos um sinal amostrado, $x_s(t)$ com espectro de frequência $X_s(j\Omega)$, podemos recuperar o sinal analógico aplicando um filtro passa-baixas, conforme ilustrado abaixo:
![RecuperacaoEpsectro](img/RecuperacaoEspectro.png)

Observe que na Figura acima considerou-se que a frequência de amostragem foi tal que não houve sobreposição do espectro (sem aliasing), observando a figura acima temos que
$$
\Omega_s - \Omega_a > \Omega_a \rightarrow \Omega_s > 2\Omega_a
$$
o que é a base para o teorema a seguir.

***Teorema de Nyquist:*** Um sinal $x_c(t)$  limitado em banda, com $X_a(j\Omega) = 0$ para $|\Omega| \geq \Omega_a$, pode ser univocamente recuperado de suas amostras $x(n)$ tomadas a uma taxa de
$$
\Omega_s \geq 2\Omega_a
$$
Observação: Esse teorema também é atribuído a <a href="https://www.britannica.com/science/information-theory/Classical-information-theory">Shannon</a> e <a href="https://en.wikipedia.org/wiki/Vladimir_Kotelnikov">Kotelnikov</a>.


<font color='green'>Exercício:</font> Seja $x_a(t) = \cos(\Omega_0 t)$, esboce o espectro de $x_s(t)$ obtido amostrando $x_a(t)$ para $\Omega_s > 2\Omega_0$ e $\Omega_s < 2 \Omega_0$.

***Espectro Analógico x Discreto:*** Lembremos que o sinal amostrado é dado por 
$x_s(t) = \sum\limits_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)$ e aplicando Fourier
$$
X_s(j\Omega) = \int_{-\infty}^{\infty} x_s(t)e^{-j\Omega t}dt = \int_{-\infty}^{\infty} \sum_{n=-\infty}^{\infty}x_c(nT_s)\delta(t - nT_s)e^{-j\Omega t}dt
$$
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty} x_c(nT_s)\int_{-\infty}^{\infty} \delta(t - nT_s)e^{-j\Omega t}dt  = \sum_{n=-\infty}^{\infty} x_c(nT_s)e^{-j\Omega T_s}  
$$
e como 
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x(n)e^{-j\omega}
$$
comparando as duas equações temos que se fizermos $\omega = \Omega T_s$ na segunda obtemos a primeira, logo $X(e^{j\omega})|_{\omega = \Omega Ts} = X_s(j\Omega)$

Até o momento trabalhamos com o sinal $x_c(t)$ do qual obtemos 
$$
X_s(j\Omega) = \sum_{n=-\infty}^{\infty}x_c(nT_s)e^{-j\Omega T_s n}
$$
entretanto, se observarmos apenas as amostras, sem considerar o período de amostragem temos
$$
x[n] = x_c(nT_s)
$$
e
$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty}x[n]e^{-j\omega n}
$$
ou seja
$$
X_s(j\Omega) = X(e^{j\omega})|_{\omega=\Omega T_s} = X(e^{j\Omega T_s})
$$
Temos que 
$$
X(e^{j\Omega T_s}) = \frac{1}{T}\sum_{k=-\infty}^{\infty}X_c[j(\Omega - k\Omega_s)]
$$
logo
$$
X(e^{j\omega}) =\frac{1}{T_s}\sum_{k=-\infty}^{\infty}X_c\left[j\left(\frac{\omega}{T_s} - \frac{2\pi k}{T_s}\right)\right]
$$


***Análise de relação entre $\Omega$ e $\omega$***

Partindo do fato que $\omega = \Omega T$, observemos que:
1. $\omega = \pi$ então $\pi = \Omega T$ e $2\pi f T = \pi$ isolando $f$ temos $ f = \frac{1}{2T} = \frac{F_s}{2}$, 

2. $\omega = 2\pi$ então $\pi = \Omega T$ que corresponde a $f = F_s$, e

3. Se $F_B$ for a maior frequência do sinal $\omega_B = 2\pi F_B T_s = \frac{2\pi F_B}{F_s}$ e como $F_s > 2F_B$, $\omega_B < \pi$.

A componente do sinal analógico que for mapeada em $\Omega_s - \Omega_B$ em $X_s(j\Omega)$ será mapeada em $\omega = 2\pi - \frac{2\pi F_B}{F_s}$ que terá valor $\pi < \omega < 2\pi$.

***Reconstrução a partir das amostras***

Partindo das amostras
$$
x_s(t) = \sum_{n=-\infty}^{\infty}x(n)\delta(t - nT_s)
$$
sendo $T_s$ o período de amostragem, e $x(n)$ o valor da amostra em $t = nT_s$.

Para reconstruir o sinal analógico a partir das amostras, podemos utilizar um filtro passa-baixas ideal $H_r(j\Omega)$, e obter na sua saída
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT_s)
$$
sendo $h_r(t)$ a resposta ao impulso do filtro passa baixas ideal, cujo módulo de resposta em frequência tem o formato
![title](img/FPB.png)
com transformada inversa
$$
h_r(t) = \frac{\sin(\pi t/T_s)}{\pi t/T_s}
$$

A reconstrução ser dada por
$$
x_r(t) = \sum_{n=-\infty}^{\infty}x(n)h_r(t - nT_s) = \sum_{n=-\infty}^{\infty}x[n]\frac{\sin[\pi(t -nT_s)/T_s]}{\pi(t - nT_s)/T_s},\ \ logo
$$
![title](img/fig49abcOppenheim2aEdicao.png)

```python
%pylab inline

t = linspace(-1,1,100) # redefine this here for convenience

fig = figure()
ax = fig.add_subplot(111) # create axis handle
k=0
fs=2 # makes this plot easier to read
ax.plot (t,sinc( k - fs * t),
         t,sinc( k+1 - fs * t),'--',k/fs,1,'o',(k)/fs,0,'o',
         t,sinc( k-1 - fs * t),'--',k/fs,1,'o',(-k)/fs,0,'o'
)
ax.hlines(0,-1,1)
ax.vlines(0,-.2,1)
ax.annotate('Valor da amostra aqui',
            xy=(0,1),
            xytext=(-1+.1,1.1),
            arrowprops={'facecolor':'red','shrink':0.05},
            )
ax.annotate('Sem interferência aqui',
            xy=(0,0),
            xytext=(-1+.1,0.5),
            arrowprops={'facecolor':'green','shrink':0.05},
            )
```

```python
fs=5.0 # sampling rate
k=array(sorted(set((t*fs).astype(int)))) # sorted coefficient list
fig,ax = subplots()

ax.plot(t,(sin(2*pi*(k[:,None]/fs))*sinc(k[:,None]-fs*t)).T,'--', # individual whittaker functions
        t,(sin(2*pi*(k[:,None]/fs))*sinc(k[:,None]-fs*t)).sum(axis=0),'k-', # whittaker interpolant
     k/fs,sin(2*pi*k/fs),'ob')# samples
ax.set_xlabel('time',fontsize=14)
ax.axis((-1.1,1.1,-1.1,1.1));
```

```python
interval=[] # piecewise domains
apprx = []  # line on domains

# build up points *evenly* inside of intervals
tp = hstack([ linspace(t[i],t[i+1],20,False) for i in range(len(t)-1) ])

# construct arguments for piecewise2
for i in range(len(t)-1):
   interval.append( np.logical_and(t[i] <= tp,tp < t[i+1]))
   apprx.append( (x[i+1]-x[i])/(t[i+1]-t[i])*(tp[interval[-1]]-t[i]) + x[i])
x_hat = np.piecewise(tp,interval,apprx) # piecewise linear approximation

ax1 = figure().add_subplot(111)
ax1.fill_between(tp,x_hat,sin(2*pi*f*tp),facecolor='red')
ax1.set_xlabel('time',fontsize=18)
ax1.set_ylabel('Amplitude',fontsize=18)
ax2 = ax1.twinx()
sqe = ( x_hat - sin(2*pi*f*tp))**2
ax2.plot(tp, sqe,'r')
ax2.axis(xmin=-1,ymax= sqe.max() )
ax2.set_ylabel('Erro quadrado', color='r',fontsize=18)
ax1.set_title('Error com a interpolacao linear')
```

```python
ax = figure().add_subplot(111)
t = linspace(-1,1,100) # redefine this here for convenience
ts = arange(-1,1+1/fs,1/fs) # sample points
num_coeffs=len(ts)
sm=0
for k in range(-num_coeffs,num_coeffs): # since function is real, need both sides
   sm+=sin(2*pi*(k/fs))*sinc( k - fs * t)
ax.plot( t,sm,'--',t,sin(2*pi*t),ts, sin(2*pi*ts),'o')
ax.set_title('Taxa de amostragem=%3.2f Hz' % fs )

# ax.figure.savefig('figure_00@.png', bbox_inches='tight', dpi=300)

ax1 = figure().add_subplot(111)
ax1.fill_between(t,sm,sin(2*pi*f*t),facecolor='red')
ax1.set_ylabel('Amplitude',fontsize=18)
ax2 = ax1.twinx()
sqe = (sm - sin(2*pi*f*t))**2
ax2.plot(t, sqe,'r')
ax2.axis(xmin=0,ymax = sqe.max())
ax2.set_ylabel('Erro ao quadrado', color='r',fontsize=18)
ax1.set_title('Error interpolando com a sinc')

# ax1.figure.savefig('figure_00@.png', bbox_inches='tight', dpi=300)
```

***Filtragem anti-aliasing***

Caso o sinal de entrada não seja limitado em banda  a taxa de Nyquist necessária é muito alta. Uma forma de reduzir esse problema é filtrar o sinal antes de amostrar.

Por exemplo, considere um sinal de voz, que a priori contém frequências na faixa audível (20Hz a 20 kHz), se aplicarmos um filtro em 4 kHz o sinal resultante ainda é inteligível e permite a identificação da pessoa que produziu a fala. Essa filtragem permite que seja usado um conversor analógico de 8 kHz por exemplo.

O filtro usado antes da amostragem é conhecido como filtro *anti-aliasing*. Como exemplo, é possível utilizar um filtro passa=baixas:
$$
H(j\Omega) = \left\{\begin{array}{c c} 1, & |\Omega| < \Omega_c  \\ 0, & |\Omega| > \Omega_c \end{array}\right.
$$
