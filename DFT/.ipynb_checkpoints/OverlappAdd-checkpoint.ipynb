```python
## Implementa a convolução vi FFT
# Versão 20/03/2024

import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([np.ones(4)])
x2 = np.ones(4)

x1 = np.append(x1,[0,0,0])
x2 = np.append(x2,[0,0,0])


#plt.title("Sinc Function")
#plt.ylabel("Sinc(n)")
#plt.xlabel("n")
plt.show()

X1_k = np.fft.fft(x1)
X2_k = np.fft.fft(x2)

X3_k = X1_k*X2_k;

x3_n = np.fft.ifft(X3_k)

plt.stem(x3_n)
```
